Соединим датасет для pool в catboost

```python
eval_pool = Pool(X_test, y_test, cat_features=cat_features)
curve = get_roc_curve(model, eval_pool)
(fpr, tpr, thresholds)=curve
roc_auc=sklearn.metrics.auc(fpr, tpr)
```

Закодирую данные с помощью OneHotEncoding и MeanTargetEncoding.

Применю OneHotEncoding, если уникальных значений меньне 5, иначе применю MeanTargetEncoding

```python
for col in categorical_columns:
    
    ### К колонкам с маленькой размерностью применим one-hot
    if data[col].nunique() < 5:
        one_hot = pd.get_dummies(data[col], prefix=col, drop_first=True)
        data = pd.concat((data.drop(col, axis=1), one_hot), axis=1)        
    ### К остальным - счетчики
    else:
        mean_target = data.groupby(col)['churn'].mean()
        data[col] = data[col].map(mean_target)
```

```python
X = data.drop('churn', axis=1)
y = data['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
C = np.linspace(0.1, 10, num=10)
```

```python
pipe = Pipeline([('scaler', StandardScaler()), 
                 ('clf', LogisticRegression(penalty='l2',
                                            max_iter=10000))])
```

```python
parameters = {'clf__C':C}
grid_clf = GridSearchCV(pipe, parameters, cv=4)
grid_clf.fit(X_train, y_train)
```

```python
clf__C = grid_clf.best_params_['clf__C']
```

```python
pipe.set_params(clf__C = clf__C)
pipe.fit(X_train, y_train)
roc_auc_clf = roc_auc_score(y, pipe.predict_proba(X)[:, 1])
print(f"ROC-AUC (LogisticRegression) = {roc_auc_clf}")
```

Напишем для SVM функцию, которая возвращает вероятности к принадлежности классу.

Для каждого числа необходимо сделать преобразование в выражение вида `P(y = +1 | x)`.

Например, вы обучили `SVM`, и на тестовых данных модель выдала следующие выходы `decision_function`:

(-10, -5, 0, +2, +10, +15)

Для каждого числа необходимо сделать преобразование в выражение вида `P(y = +1 | x)`.

С одной стороны, отрицательный знак числа будет сигнализировать нам о том, что `P(y = +1 | x) < 0.5`.

Тогда положительный о том, что `P(y = +1 | x) > 0.5`. 

С другой стороны, для тех объектов, в которых модель уверена больше всего, положим краевые вероятности. Для примера выше:

`P(y = +1 | -10) = 0`, `P(y = +1 | +15) = 1`. Для всех промежуточных объектов применим пропорциональное преобразование. Например:

$$
P(y = +1 | -5) = \frac{|-5-(-10)|}{|-10|} \cdot 0.5
$$

$$
P(y = +1 | +2) = \frac{|+2|}{|+15|} \cdot 0.5 + 0.5
$$
Таким образом мы можем перевести выходы SVM в отрезок [0;1] и попробовать интерпретировать новые числа как вероятности, чтобы изобразить калибровочную кривую. При таком преобразовании пропорции внутри первоначально отрицательных и внутри положительных чисел сохраняются.

```python
def proba(distance):
    mx_dist = max(distance)
    mn_dist = min(distance)
    
    func = lambda x:  (abs(x - mn_dist)/abs(mn_dist))*0.5 if x < 0 else (abs(x)/abs(mx_dist))*0.5 + 0.5
    
    return np.array(list(map(func, distance)))
```

```python
distance = pipe.decision_function(X_test)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import auc 

distance = pipe.decision_function(X_test)
fpr, tpr, thresholds = roc_curve(y_test, proba(distance))
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
print(roc_auc)
display.plot()
plt.show()
```

```python
rfr__n_estimators = list(map(int, np.linspace(80, 121, num=9)))

pipe = Pipeline([('scaler', StandardScaler()), 
                 ('rfr', RandomForestClassifier())])
```

```python
parameters = {'rfr__n_estimators':rfr__n_estimators}
grid_svc = GridSearchCV(pipe, parameters, cv=4)
grid_svc.fit(X_train, y_train)
```

```python
rfr__n_estimators = grid_svc.best_params_['rfr__n_estimators']
```

```python
pipe.set_params(rfr__n_estimators = rfr__n_estimators)
pipe.fit(X_train, y_train)
roc_auc_rfr = roc_auc_score(y_test, pipe.predict_proba(X_test)[:, 1])
print(f"ROC-AUC (RandomForestRegressor) = {roc_auc_rfr}")
```
